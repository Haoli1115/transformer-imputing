# Data loading and information extraction

```python
# Import packages
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import datetime
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
import plotly.express as px
import plotly.graph_objects as go
plt.style.use("seaborn")
```

```python
#Define colormap for plotting
cmap=sns.diverging_palette(0, 125, s=80, l=50, n=50)
```

```python
#Load data
years=[2015,2016,2017,2018,2019,2020]
all_data=pd.read_csv("../data/all_data_processed.csv", index_col=0)
#Create datetime index
all_data.index = pd.to_datetime(all_data.index)
countries = all_data["Country"].unique()
feature_names = all_data.columns[2:]
no_features = len(all_data.columns[2:])
```

## Extract n/e dictionary

```python
# Extract n/e dictionary
missing_dict = dict()
for country in countries:
    #Get completely missing features for a country
    country_list = (all_data.loc[all_data["Country"]==country] == "n/e").sum()
    country_list = country_list[country_list > 0].index.to_list()
    missing_dict[country] = country_list
```

```python
#Replace n/e with 0 for production and set to float
all_data.replace({"n/e":0}, inplace = True)
all_data.iloc[:,4:] = all_data.iloc[:,4:].astype("float64")
```

# Whole Missing value matrix

```python
#Prepare count
data_grouped=all_data.groupby(["Country"]).count()
#Count occurences per year
data_count=all_data.groupby(["Country"]).size()
data_missing=data_grouped.copy().drop("Year",axis=1)
#Calculate percentage of missing values respective to all possible available data
data_missing = data_missing.apply(axis=0, func = lambda x: x/data_count)
data_missing = 1-data_missing
data_missing.insert(0, "Country availability", 1 - data_count/data_count.max())

#Insert n/e
for country in countries:
    data_missing.loc[country, missing_dict[country]] = "n/e"
    
#Save to excel sheet
data_missing.to_excel("missing_data_matrix.xlsx")
data_missing.head()
```

## Aggregate and rename features

```python
#Other
other = ["Other", "Fossil Peat", "Other renewable", "Waste", "Fossil Oil", "Fossil Coal-derived gas", "Geothermal"]
all_data["Other"] = all_data[other].sum(axis=1, skipna = False)
all_data.drop(other[1:],axis=1, inplace=True)

#Hydro
hydro = ["Hydro Run-of-river and poundage", "Hydro Water Reservoir"]
all_data["Hydro"] = all_data[hydro].sum(axis=1, skipna = False)
all_data.drop(hydro, axis = 1, inplace = True)

#Rename
all_data.rename(columns = {"Fossil Brown coal/Lignite":"Lignite", "Solar":"PV",
                           "Fossil Hard coal": "Hard coal",
                           "Hydro Pumped Storage":"Pumped Storage"}, inplace=True)
```

## Smaller missing value matrix

```python
#Prepare count
data_grouped=all_data.groupby(["Country"]).count()
#Count occurences per year
data_count=all_data.groupby(["Country"]).size()
data_missing=data_grouped.copy().drop("Year",axis=1)
#Calculate percentage of missing values respective to all possible available data
data_missing = data_missing.apply(axis=0, func = lambda x: x/data_count)
data_missing = 1-data_missing
data_missing.insert(0, "Country availability", 1 - data_count/data_count.max())

#Insert n/e
for country in countries:
    missing_features = missing_dict[country]
    for feature in missing_features:
        if (feature in data_missing.columns) & (feature !="Other"):
            data_missing.loc[country, feature] = "n/e"

#Save to excel sheet
data_missing.to_excel("missing_data_matrix_subset.xlsx")
data_missing.head()
```

# Analyze gaps of missing values
We create a function that takes the data, a country and a feature and returns the number of gaps, as well as the respective gap length

```python
def get_gaps(data, country, feature):
    #Filter country and feature
    filtered_data=data[data["Country"]==country].copy()[[feature]]
    gap_data = filtered_data.loc[filtered_data[feature].isna()]
    #Calculate percentage of missing values
    perc_missing = np.round(len(gap_data)/len(filtered_data),4)
    
    #Calculate gap length and number of gaps
    n_gaps = 0
    gap_lengths = []
    consecutive = 0
    index = gap_data.index
    #Create loop
    for cnt,ind in enumerate(index):
        consecutive+=1
        if cnt<(len(index)-1) and index[(cnt+1)] == ind+datetime.timedelta(hours=1):
            n_gaps+=0
        else:
            gap_lengths.append(consecutive)
            consecutive = 0
            n_gaps+=1
    return (n_gaps,gap_lengths,perc_missing)
```

## Example of analyzing gaps for Day ahead price

```python
feature = "Day_ahead_price"
#countries already defined
results = pd.DataFrame(columns=("n_gaps","Gaps","% missing"))
all_gaps = [] 

#Iterate over all countries
for country in countries:
    n_gaps, gaps, perc_missing = get_gaps(all_data,country,feature)
    results.loc[country] = [n_gaps, gaps, perc_missing]
    all_gaps.extend(gaps)
```
